```python
id = []
address = []
city = []
state = []
zip_code = []
price = []
estimated_rent = []
bedrooms = []
bathrooms = []
square_feet = []
description = []

for result in result_items:
    # id
    id.append(result.get('id'))
    # address
    address.append(result.get('address'))
    # city
    city.append(result.get('addressCity'))
    # state
    state.append(result.get('addressState'))
    # zip code
    zip_code.append(result.get('addressZipcode'))
    # price
    price.append(result.get('unformattedPrice'))
    # rent
    estimated_rent.append(result.get('hdpData', {}).get('homeInfo', {}).get('rentZestimate'))
    # bedrooms
    bedrooms.append(result.get('beds'))
    # bathrooms
    bathrooms.append(result.get('baths'))
    # square footage
    square_feet.append(result.get('area'))
    # description
    description.append(result.get('detailUrl'))
```

```python
df_homes = pd.DataFrame({'id': id, 'address': address, 'city': city, 'state':state, 'zip_code': zip_code, 'price': price, 'estimated_rent': estimated_rent, 'bedrooms': bedrooms, 'bathrooms': bathrooms, 'square_feet': square_feet, 'description': description})
```
